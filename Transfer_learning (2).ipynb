```
from keras.preprocessing import image
import matplotlib.pyplot as plt
import numpy as np
```

```
img = image.load_img("/content/drive/My Drive/Pokemon_dataset/val_images/Abra/d1ba8548fced4b53821bfc1e0612324f.jpg",target_size=(224,224))
```

```
#need to convert the image to float values for plt.imshow by diving the image array by /255.
x = image.img_to_array(img)
x = x/255. #import to visulaize images
```

```
#importing libraries
from keras.applications import ResNet50
from keras.applications.resnet50 import preprocess_input,decode_predictions
```

```
# Implementing data augmentation and using data generator 
train_data = image.ImageDataGenerator(rotation_range=10,width_shift_range=0.02,height_shift_range=0.02,shear_range=0.02,zoom_range=0.05,rescale=1./255,horizontal_flip=True,vertical_flip=True)

val_data  = image.ImageDataGenerator(rescale = 1./255)
```

```
#data genarators 
train_data_gen = train_data.flow_from_directory(data_path,target_size=(100,100),batch_size=32,class_mode='categorical',shuffle=True)
val_data_gen = val_data.flow_from_directory(val_path,target_size=(100,100),batch_size=32,class_mode='categorical',shuffle=True)
```

```
x,y = train_data_gen.next()
```

```
#Models

from keras.applications import ResNet50
from keras.applications.resnet50 import decode_predictions
from keras.layers import *
from keras.models import Model
```

```
model = ResNet50(input_shape = (224,224,3),include_top=False,weights = "imagenet")
```

```
model.summary()
```

```
av1 = GlobalAveragePooling2D()(model.output)
fc1 = Dense(512,activation = 'relu')(av1)
d1 = Dropout(0.2)(fc1)
fc2 = Dense(256,activation = 'relu')(d1)
d2 = Dropout(0.2)(fc2)
fc3 = Dense(149,activation='softmax')(d2)
```

```
model_new = Model(input=model.input,output=fc3)

model_new.summary()
```

```
for ix in range(len(model_new.layers)):
  print(ix,model_new.layers[ix])
```

```
#keeping the top layers of resnet as non trainable and training only the classifier and last 2 conv layers of resnet50 so that model can adapt the change in dataset
for ix in range(171):
  model_new.layers[ix].trainable = False
```

```
model_new.summary()
```

```
#using Adams as optimizers with very low learning rate so to avoid sudden changes in resnet layers till where error will be backpropogated
from keras.optimizers import Adam
```

```
adam = Adam(lr = 0.009)
```

```
#adding callbacks and early stopping
from keras.callbacks import ModelCheckpoint, EarlyStopping
```

```
checkpoint = ModelCheckpoint("/content/drive/My Drive/Pokemon_dataset/pokemon_model3.0.hdf5",monitor = 'val_acc',save_best_only=True)
callback = EarlyStopping(monitor='val_acc',min_delta=0.05,patience=8,restore_best_weights=True)
```

```
model_new.compile(loss = 'categorical_crossentropy',optimizer=adam,metrics=['accuracy'])
```

```
hist = model_new.fit_generator(train_data_gen,steps_per_epoch=350,epochs=5,callbacks=[checkpoint,callback],validation_data=val_data_gen,validation_steps=50)
```

```
from keras.layers import *
```

```
from keras.models import Sequential
```

```
new_model = Sequential()

```

```
new_model = Sequential()

new_model.add(Conv2D(64,(3,3),activation='relu',input_shape=(100,100,3)))
new_model.add(MaxPool2D(2,2))
new_model.add(BatchNormalization())

new_model.add(Conv2D(128,(3,3),activation='relu'))
new_model.add(MaxPool2D(2,2))
new_model.add(BatchNormalization())

new_model.add(Conv2D(256,(3,3),activation='relu'))
new_model.add(MaxPool2D(2,2))
new_model.add(BatchNormalization())


new_model.add(Conv2D(512,(3,3),activation='relu'))
new_model.add(MaxPool2D(2,2))
new_model.add(BatchNormalization())


new_model.add(Conv2D(1028,(3,3),activation='relu'))
new_model.add(MaxPool2D(2,2))
new_model.add(BatchNormalization())

new_model.add(Flatten())

new_model.add(Dense(512,activation='relu'))
new_model.add(Dropout(0.3))


new_model.add(Dense(256,activation='relu'))
new_model.add(Dropout(0.3))


new_model.add(Dense(149,activation='softmax'))

new_model.summary()


```

```
new_model.compile(optimizer='rmsprop',loss='categorical_crossentropy',metrics=['accuracy'])
```

```
hist = new_model.fit_generator(train_data_gen,steps_per_epoch=400,epochs=30,callbacks=[checkpoint],validation_data=val_data_gen,validation_steps=50)
```
